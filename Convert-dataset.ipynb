```python
import json
import pandas as pd
```

```python
data = []

for line in open('/content/drive/MyDrive/Soft Computing Project/Cell_Phones_and_Accessories_5.json','r') :
    data.append(json.loads(line))
```

```python
print(data[0])
df = pd.DataFrame(data)
print(len(data))
```

```python

df = df.drop(columns=['reviewerName'])
df1 = df.rename(columns = {'overall': 'rating', 'asin': 'productID'}, inplace = False)

```

```python
df1.dropna(axis = 0, how ='any',inplace=True) 
df1.drop_duplicates(subset=['rating','reviewText'],keep='first',inplace=True)
```

```python
print(len(df1))
df1.to_csv('Dataset.csv')
```
